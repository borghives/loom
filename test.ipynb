```python
from loom.info import Persistable, declare_persist_db
from typing import Type
from loom.info.model import TimeInserted, TimeUpdated

@declare_persist_db(db_name="test", collection_name="test")
class TestUpdateModel(Persistable):
    model_id: ObjectId
    name: str
    value: int
    created_at: TimeInserted
    updated_time: TimeUpdated
```

```python
from typing import Annotated
from bson.objectid import ObjectId
import rich
from typing import get_origin, get_args, Union

def get_base_type(annotation):
    """Recursively unwraps Annotated and Optional/Union types."""
    origin = get_origin(annotation)
    args = get_args(annotation)

    # Handle Annotated: Extract the first argument (the actual type)
    if origin is Annotated:
        return get_base_type(args[0])
    elif origin is Union:
        return get_base_type(args[0])

    return annotation

def is_objectid(annotation, field_type: Type):
    return get_base_type(annotation) is field_type

for key, field in TestUpdateModel.model_fields.items():
    annotation = field.annotation
    if is_objectid(annotation, ObjectId):
        rich.print(field)
        print(f"{key} {field.alias} is an ObjectId field!")
```
